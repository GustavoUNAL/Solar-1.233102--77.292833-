## Leer CSV de resultados del análisis

```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('../../Data/Processed/cofactores/resultados.csv')
# listar el nombre de las columnas
print(df.columns)
print(df.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# sns.set
sns.set(style="whitegrid", context="notebook")
cols = ['GHI',
       'Producción_fotovoltaica_SFV', 'clearsky_ghi', 'P_cloud_opacity_temp',
       'Potencia', 'Date', 'HBS', 'E_SFV', 'E_solacast', 'error', 'alpha',
       'beta']
sns.pairplot(df[cols], size=2)
plt.show()
```

```python
# agrupar columna fecha por mes 
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

# graficar alpha y beta por día 
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['alpha'])
plt.plot(df['Date'], df['beta'])
plt.xlabel('Fecha')
plt.ylabel('Coeficiente')
plt.title('Coeficientes por día')
plt.legend(['alpha', 'beta'])
plt.show()
```

```python
# graficar error por día
plt.figure(figsize=(15, 3))
plt.scatter(df['Date'], df['error'])
plt.xlabel('Fecha')
plt.ylabel('Error')
plt.title('Error por día')
plt.show()
```

```python
plt.figure(figsize=(15, 3))
plt.plot(df['Date'], df['E_SFV']) 
plt.plot(df['Date'], df['E_solacast'])
plt.xlabel('Fecha')
plt.ylabel('Energía producida')
plt.title('Energía producida por día')
plt.legend(['Energía producida', 'Energía pronosticada'])
plt.show()

```
