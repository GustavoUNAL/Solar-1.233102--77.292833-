# An√°lisis de datos SFV

```python
import numpy as np

# Ruta al archivo .npy
archivo_tensor_entrenamiento_SFV = '../RNA/data_training/tensor_SFV.npy'

# Cargar el tensor
tensor_cargado_entrenamiento_SFV = np.load(archivo_tensor_entrenamiento_SFV)

# Mostrar el tensor cargado
print(tensor_cargado_entrenamiento_SFV.shape)
```

# imprimir todos los valores del tensor

```python
# imprimir todos los valores del tensor
# print(tensor_cargado_entrenamiento_SFV[0][:,0])
# .sum() 
produccionDia = tensor_cargado_entrenamiento_SFV[0][:,0]

#*(5/60) 
print(produccionDia.sum()*(5/60))
print(produccionDia.sum())
print(produccionDia.sum()*(5/60)*16.200)
```

```python
# encontrar produccionDia = tensor_cargado_entrenamiento_SFV[0][:,0] para todos los dias
produccionDias = []
for i in range(0, tensor_cargado_entrenamiento_SFV.shape[0]):
    produccionDia = tensor_cargado_entrenamiento_SFV[i][:,0]
    produccionDias.append(produccionDia.sum())

print(produccionDias)
```

```python
# buscar en produccionDias 4.11711802802009 y obtener el indice
indice = produccionDias.index(58.31908018867925)
print(indice)
```

```python
import matplotlib.pyplot as plt
from scipy.integrate import simps


# imprime el primer elemento del tensor
print(tensor_cargado_entrenamiento_SFV[0].shape)
plt.plot(tensor_cargado_entrenamiento_SFV[0][:,0])
plt.show()

# area bajo la curva de tensor_cargado_entrenamiento_SFV[0]
area = simps(tensor_cargado_entrenamiento_SFV[0][:,0])
print("Area bajo la curva de tensor_cargado_entrenamiento_SFV[0]:", area)
```

### Comparaci√≥n con la DB

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Cargar los datos del archivo CSV
df_SFV = pd.read_csv('../Data/Cleaned/SFV_2023.csv', sep=',')
# print(df_SFV.head())
# fiultrar Producci√≥n_fotovoltaica_SFV: 2023-01-02

df_SFV = df_SFV[df_SFV['Date'] == '2023-03-21']

# Graficar Producci√≥n_fotovoltaica_SFV
plt.plot(df_SFV['Producci√≥n_fotovoltaica_SFV'])
plt.show()

# suma df_SFV['Producci√≥n_fotovoltaica_SFV']
print("valor real Wh:")
print(df_SFV['Producci√≥n_fotovoltaica_SFV'].sum())

# imprime el valor maximo de df_SFV['Producci√≥n_fotovoltaica_SFV']
print("valor maximo Wh:")
maximo = df_SFV['Producci√≥n_fotovoltaica_SFV'].max()
print(maximo)
```

```python
print("valor real Wh:")
print(maximo*58.31908018867925)
print("valor de la predicci√≥n Wh:")
print(maximo*54.759453)
```

```python
# normalizar df_SFV['Producci√≥n_fotovoltaica_SFV']
scaler = MinMaxScaler()
df_SFV['Producci√≥n_fotovoltaica_SFV'] = scaler.fit_transform(df_SFV[['Producci√≥n_fotovoltaica_SFV']])
# print(df_SFV.head())
# suma df_SFV['Producci√≥n_fotovoltaica_SFV']
print("valor normalizado:")
print(df_SFV['Producci√≥n_fotovoltaica_SFV'].sum())
```

## Datos normalizados

```python
# normalizar la producci√≥n fotovoltaica
scaler = MinMaxScaler()
df_SFV['Producci√≥n_fotovoltaica_SFV'] = scaler.fit_transform(df_SFV['Producci√≥n_fotovoltaica_SFV'].values.reshape(-1,1))

# area bajo la curva de Producci√≥n_fotovoltaica_SFV
area = simps(df_SFV['Producci√≥n_fotovoltaica_SFV'])
print("Area bajo la curva de Producci√≥n_fotovoltaica_SFV:", area)
```
