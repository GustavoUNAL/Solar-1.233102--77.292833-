```python
import numpy as np

# Cambia 'ruta_al_archivo.npy' por la ruta real a tu archivo .npy
archivo = 'tensor_entradas.npy'

# Cargar el tensor
tensor = np.load(archivo)

# Mostrar información básica sobre el tensor
print("Forma del tensor:", tensor.shape)
print("Tipo de datos del tensor:", tensor.dtype)
print("Primeros elementos del tensor:", tensor.flatten()[:10])

# Realizar un análisis básico (ejemplo: promedio, máximo, mínimo)
promedio = np.mean(tensor)
maximo = np.max(tensor)
minimo = np.min(tensor)

print(f"Promedio: {promedio}, Máximo: {maximo}, Mínimo: {minimo}")

# Si el tensor es de dos dimensiones, podrías visualizarlo con matplotlib
if tensor.ndim == 2:
    import matplotlib.pyplot as plt
    plt.imshow(tensor, cmap='gray')  # Asumiendo que es una imagen en escala de grises
    plt.title("Visualización del tensor")
    plt.colorbar()
    plt.show()

```
